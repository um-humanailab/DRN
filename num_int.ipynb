```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random numbers from a normal distribution
mean = 0
std_dev = 1
size = 1000
samples = np.random.normal(mean, std_dev, size)

# Plot a histogram and KDE using seaborn
sns.histplot(samples, bins=30, kde=True, color='blue', label='Histogram and KDE')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram and KDE of Normal Distribution using Seaborn')
plt.legend()
plt.show()
```
